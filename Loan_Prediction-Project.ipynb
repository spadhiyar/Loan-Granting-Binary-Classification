# Data Cleaning

1. Loan[Loan['Credit_Score']>800] devided it by 10 
2. Covert 'Money debt' to float & remove '$' & ',' sign from column and convert to float
3. Convert 'Maximum open credit' from object to float
4. In Purpose column combined 'other' & 'Other' 
5. Years_in_current_job has '<', n/a & 'years' so will convert it into numbers & datatype will be float
6. In Home_Ownership replcae combined Home Mortage & HomeMortage
7. In 'Current_Loan_Amount' replace 99999999 with 0

```python
# Loan[Loan['Credit_Score']>800] devided it by 10

Loan.loc[Loan['Credit_Score']>800,'Credit_Score'] = Loan.loc[Loan['Credit_Score']>800,'Credit_Score']/10
```

```python
Loan.describe()
```

```python
# Covert 'Money debt' to float & remove '$' & ',' sign from column and convert to float

Loan['Monthly_Debt']= Loan['Monthly_Debt'].str.replace(',','').map(lambda x: x[1:]).astype('float')
```

```python
Loan.Monthly_Debt.head()
```

```python
# Convert 'Maximum open credit' from object to float

Loan['Maximum_Open_Credit']= Loan['Maximum_Open_Credit'].replace('#VALUE!',0).astype('float64')
```

```python
Loan.Maximum_Open_Credit.head()
```

```python
# convert Current_Credit_Balance into float

Loan['Current_Credit_Balance'] = Loan['Current_Credit_Balance'].astype('float64')
```

```python
Loan.Current_Credit_Balance.head()
```

```python
# In Purpose column combined 'other' & 'Other'

Loan.loc[Loan['Purpose']=='other','Purpose'] = 'Other'
```

```python
Loan.Years_in_current_job.value_counts()
```

```python
Loan['Years_in_current_job'] = Loan['Years_in_current_job'].str.strip('+ years')
Loan['Years_in_current_job'] = Loan['Years_in_current_job'].replace('< 1', 0)
Loan['Years_in_current_job'] = Loan['Years_in_current_job'].replace('n/', 0)
Loan['Years_in_current_job'] = Loan['Years_in_current_job'].astype('float64')
```

```python
Loan.Years_in_current_job.value_counts()
```

```python
Loan.Home_Ownership.value_counts()
```

```python
Loan['Home_Ownership']=Loan['Home_Ownership'].replace('HaveMortgage','Home Mortgage')
```

```python
Loan.Home_Ownership.value_counts()
```

```python
Loan.isnull().sum()
```

```python
Loan['Current_Loan_Amount']=Loan['Current_Loan_Amount'].replace(99999999,0)
```

# Cross Validation

```python
from sklearn.model_selection import KFold #for K-fold cross validation
from sklearn.model_selection import cross_val_score #score evaluation
from sklearn.model_selection import cross_val_predict #prediction
kfold = KFold(n_splits=10, random_state=22) # k=10, split the data into 10 equal parts
xyz=[]
accuracy=[]
std=[]
classifiers=['Logistic Regression','Random Forest','Gradient Boosting']
models=[LogisticRegression(),RandomForestClassifier(n_estimators=100),GradientBoostingClassifier()]
for i in models:
    model = i
    cv_result = cross_val_score(model,X,Y, cv = kfold,scoring = "accuracy")
    cv_result=cv_result
    xyz.append(cv_result.mean())
    std.append(cv_result.std())
    accuracy.append(cv_result)
new_models_dataframe2=pd.DataFrame({'CV Mean':xyz,'Std':std},index=classifiers)       
new_models_dataframe2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

new_models_dataframe2['CV Mean'].plot.barh(width=0.8)
plt.title('Average CV Mean Accuracy')
fig=plt.gcf()
fig.set_size_inches(8,5)
plt.show()
```

```python
f,ax=plt.subplots(3,2,figsize=(12,10))

y_pred = cross_val_predict(LogisticRegression(),X,Y,cv=10)
sns.heatmap(confusion_matrix(Y,y_pred),ax=ax[0,0],annot=True,fmt='2.0f')
ax[0,0].set_title('Matrix for Logistic Regression')

y_pred = cross_val_predict(RandomForestClassifier(n_estimators=100),X,Y,cv=10)
sns.heatmap(confusion_matrix(Y,y_pred),ax=ax[1,0],annot=True,fmt='2.0f')
ax[1,0].set_title('Matrix for Randomforest')

y_pred = cross_val_predict(GradientBoostingClassifier(),X,Y,cv=10)
sns.heatmap(confusion_matrix(Y,y_pred),ax=ax[2,0],annot=True,fmt='2.0f')
ax[2,0].set_title('Matrix for Gradientboosting')
```
